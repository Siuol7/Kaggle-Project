```python
heart=pd.read_csv('heart.csv')
heartdf=pd.DataFrame(heart)
heartdf
```

```python
heartdf.isnull().sum()
```

```python
heartdf.corr()
plt.figure(figsize=(12,12))
sns.heatmap(heartdf.corr(),annot=True,cmap='coolwarm')
```

```python
heartdf['sex']=np.where(heartdf['sex']==1,'Male','Female')
heartdf.loc[heartdf['cp']==0,'cp']='Typical Agina'
heartdf.loc[heartdf['cp']==1,'cp']='Atypical Agina'
heartdf.loc[heartdf['cp']==2,'cp']='Non-Aginal Pain'
heartdf.loc[heartdf['cp']==3,'cp']='Asymptomatic'
heartdf['fbs']=np.where(heartdf['fbs']==0,'False','True')
heartdf['exang']=np.where(heartdf['exang']==1,'Yes','No')
heartdf.loc[heartdf['slope']==0,'slope']='Unsloping'
heartdf.loc[heartdf['slope']==1,'slope']='Flat'
heartdf.loc[heartdf['slope']==2,'slope']='Downsloping'
heartdf.loc[heartdf['thal']==1,'thal']='Normal'
heartdf.loc[heartdf['thal']==2,'thal']='Fixed Defect'
heartdf.loc[heartdf['thal']==3,'thal']='Revesible Defect'
heartdf['target']=np.where(heartdf['target']==1,'Heart Disease','No Heart Disease')

heartdf



```

```python
heartdf.columns=heartdf.columns.str.capitalize()
heartdf.rename(columns={'Sex':'Gender'},inplace=True)
heartdf
```

```python
 #Genral Information
 Par=heartdf['Age'].count()
 print('Number of Participants:',Par)

 Ha=heartdf['Age'][heartdf['Target']=='Heart Disease'].count()
 print('Number of Heart Attack Patients:',Ha)

 Average_age=round(heartdf['Age'].mean(),2)
 print('Average age of Sample:',Average_age)
```

```python
#Sex Distribution
SD=heartdf.groupby('Gender')['Gender'].count().reset_index(name='Number of people')
SD=pd.DataFrame(SD)
explode=[0.1,0]
plt.figure(figsize=(7,7))
plt.pie(SD['Number of people'],labels=SD['Gender'],explode=explode,shadow=True, colors=['Pink','maroon'],autopct='%0.1f%%')
plt.legend(SD['Gender'],loc='best')
plt.show()
```

```python
#Heart Attack Distribution
HA=heartdf.groupby('Target')['Target'].count().reset_index(name='Number of people')
HA=pd.DataFrame(HA)
expldoe=[0.1,0]
plt.figure(figsize=(7,7))
plt.pie(HA['Number of people'],labels=HA['Target'],colors=['Maroon','Pink'],explode=explode,shadow=True,autopct='%0.1f%%')
plt.legend(HA['Target'],loc='best')
plt.show()
```

```python
# Age Distribution
AD = heartdf.groupby(['Age', 'Target'])['Target'].count()
AD.unstack().plot(color=['maroon','pink'])
plt.xticks(rotation=0)
plt.ylabel('Number of People')
plt.rcParams["figure.figsize"] = (10, 5)
plt.legend(loc='best')
plt.show()
```

```python
#Age and Gender Distribution
AGD=heartdf.groupby(['Age','Gender'])['Gender'].count()
AGD.unstack().plot(kind='bar',color=['maroon','pink'])
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.legend(loc='best')
plt.rcParams["figure.figsize"] = (20, 5)

```
