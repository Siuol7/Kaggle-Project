# **Data Wrangling**

```python
train=pd.read_csv('train.csv')
df_train=pd.DataFrame(train)
df_train
```

```python
test=pd.read_csv('test.csv')
df_test=pd.DataFrame(test)
df_test
```

```python
#Function split numerical and string values
def extract_numerical_v2(value):
    if isinstance(value, str):
        numeric_part = ''.join(filter(str.isdigit, value.split()[0]))
        return float(numeric_part) / 10 if numeric_part else None  # Return None if the numeric part is empty
    return value
```

```python
num_feature=['back_legroom', 'front_legroom','fuel_tank_volume','height','length','wheelbase','width']
for feature in num_feature:
    df_train[feature] = df_train[feature].apply(extract_numerical_v2)
    df_test[feature] = df_test[feature].apply(extract_numerical_v2)
```

```python
#Re-check
df_train[num_feature].head(5)
```

```python
#Function to extract horsepower from 'power' column
def extract_horsepower(value):
    if isinstance(value, str) and "hp" in value:
        return float(value.split("hp")[0].strip())
    return None


#Function to extract RPM at which horsepower is achieved
def extract_horsepower_rpm(value):
    if isinstance(value, str) and "RPM" in value:
        return float(value.split("@")[1].split("RPM")[0].replace(",", "").strip())
    return None

```

```python
#Apply functions to extract 'horsepower' and 'horsepower_rpm' from 'power' column
df_train['horsepower'] = df_train['power'].apply(extract_horsepower)
df_train['horsepower_rpm'] = df_train['power'].apply(extract_horsepower_rpm)
df_test['horsepower'] = df_test['power'].apply(extract_horsepower)
df_test['horsepower_rpm'] = df_test['power'].apply(extract_horsepower_rpm)
df_train.drop(columns=['power'],inplace=True)
df_test.drop(columns=['power'],inplace=True)
```

```python
#Function to extract torque from 'torque' column
def extract_max_torque(value):
    if isinstance(value, str) and "lb-ft" in value:
        return float(value.split("lb-ft")[0].strip())
    return None

#Function to extract RPM at which torque is achieved
def extract_torque_rpm(value):
    if isinstance(value, str) and "RPM" in value:
        return float(value.split("@")[1].split("RPM")[0].replace(",", "").strip())
    return None
```

```python
#Apply functions to extract 'torque' and 'torque_rpm' from 'torque' colummn
df_train['max_torque'] = df_train['torque'].apply(extract_max_torque)
df_train['torque_rpm'] = df_train['torque'].apply(extract_torque_rpm)
df_test['max_torque'] = df_test['torque'].apply(extract_max_torque)
df_test['torque_rpm'] = df_test['torque'].apply(extract_torque_rpm)
df_train.drop(columns=['torque'],inplace=True)
df_test.drop(columns=['torque'],inplace=True)
```

```python

#Identify numerical, categorical, and boolean columns in the TRAINING dataset
numerical_cols_train = df_train.select_dtypes(include=['float64', 'int64']).columns
categorical_cols_train = df_train.select_dtypes(include=['object']).columns
boolean_cols_train = df_train.select_dtypes(include=['bool']).columns
```

```python
#Identify numerical, categorical, and boolean columns in the TÉTING dataset
numerical_cols_test = df_test.select_dtypes(include=['float64', 'int64']).columns
categorical_cols_test = df_test.select_dtypes(include=['object']).columns
boolean_cols_test = df_test.select_dtypes(include=['bool']).columns
```

```python
#Adjusting the columns list to only consider columns that are present in both train and test datasets
common_numerical_cols = [col for col in numerical_cols_train if col in df_test.columns]
common_categorical_cols = [col for col in categorical_cols_train if col in df_test.columns]
common_boolean_cols = [col for col in boolean_cols_train if col in df_test.columns]
```

```python
#Impute numerical columns with median
for col in common_numerical_cols:
    median_val_train = df_train[col].median()
    df_train[col].fillna(median_val_train, inplace=True)
    df_test[col].fillna(median_val_train, inplace=True)  # Using median from train data for test data
#Impute categorical and boolean columns with mode
for col in common_categorical_cols + common_boolean_cols:
    mode_val_train = df_train[col].mode()[0]
    df_train[col].fillna(mode_val_train, inplace=True)
    df_test[col].fillna(mode_val_train, inplace=True)  # Using mode from train data for test data
```

```python
#Re-checking missing values
total_nan=df_train.isnull().sum().sum()
total_nan
```

```python
def transform_and_encode(column):
    # Identify rows with multiple values (assuming comma indicates multiple values) and set them to 'other'
    column[column.str.contains(',', na=False)] = 'other'

    #If there are more than 5 unique values, keep top 5 and set rest to 'other'
    if column.nunique() > 5:
        top_5_values = column.value_counts().head(5).index.tolist()
        column[~column.isin(top_5_values)] = 'other'

    return column

#Transform and encode the categorical columns
for col in common_categorical_cols:
    df_train[col] = transform_and_encode(df_train[col])
    df_test[col] = transform_and_encode(df_test[col])

#Perform one-hot encoding
df_train = pd.get_dummies(df_train, columns=common_categorical_cols, drop_first=False)
df_test = pd.get_dummies(df_test, columns=common_categorical_cols, drop_first=False)



df_train.head()
```

# **Building Model**

```python
from sklearn.preprocessing import StandardScaler
Scale=StandardScaler()

#Feature Scaling: Standardize numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#List of numerical features to scale
features_to_scale = common_numerical_cols + ['max_torque', 'torque_rpm', 'horsepower',
                                            'horsepower_rpm']

#Apply scaling on the train and test datasets
df_train[features_to_scale] = scaler.fit_transform(df_train[features_to_scale])
df_test[features_to_scale] = scaler.transform(df_test[features_to_scale])  # Use the same scaler fit from train data



#Ensure Data Consistency Between Train and Test: Align columns
missing_cols_train = set(df_train.columns) - set(df_train.columns)
missing_cols_test = set(df_test.columns) - set(df_test.columns)

for col in missing_cols_train:
    df_train[col] = 0

for col in missing_cols_test:
    df_test[col] = 0


df_train.head(), df_test.head()
```

```python
#Train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
lr=LinearRegression()
x_data= df_train.loc[:, df_train.columns != 'price']
y_data=df_train['price']
x_train, x_test, y_train, y_test= train_test_split(x_data,y_data,test_size= 0.3, random_state=0)
```

### **Multiple Linear Regression**

```python
#Multiple Linear Regression
lr.fit(x_train,y_train)
y_pred_mlr=lr.predict(x_test)
```

### **Ridge Regression**

```python
#GridSearchCV->alpha Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
parameters=[{'alpha':[0.01,0.1,1,10,100]}]
RR=Ridge()
Grid=GridSearchCV(RR, parameters, cv=4)
Grid.fit(x_data,y_data)
Grid.best_estimator_

```

```python
#RdigeRegression
RidgeModel=Ridge(alpha=10)
RidgeModel.fit(x_train,y_train)
y_pred_rr=RidgeModel.predict(x_test)
```

### **Random Forest**

```python
#GridSearch to find the best n_estimator for random forest

from sklearn.ensemble import RandomForestRegressor

#Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300], # Number of trees in random forest
    'max_features': ['auto', 'sqrt'], # Number of features to consider at every split
    'max_depth': [10, 20, 30], # Maximum number of levels in tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4], # Minimum number of samples required at each leaf node
}

#Create the base model to tune
rf = RandomForestRegressor(random_state=42)

#Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

#Fit the grid search to the data
grid_search.fit(x_train, y_train)

#Best estimator
best_grid = grid_search.best_estimator_

#Use the best estimator to make predictions
y_pred = best_grid.predict(x_test)

#If you want to check the feature importances,
#which shows how much each feature contributes to the prediction
feature_importances = best_grid.feature_importances
print(f"Feature importances: {feature_importances}")

```

```python
#Best parameters found
print(grid_search.best_params_)
```

```python
# RandomForestModel
rf = RandomForestRegressor(n_estimators=200, random_state=0)

# Huấn luyện mô hình trên dữ liệu huấn luyện
rf.fit(x_train, y_train)

# Dự đoán kết quả trên dữ liệu kiểm tra
y_pred_rf = rf.predict(x_test)
```
